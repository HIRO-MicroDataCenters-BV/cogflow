```python
import kfp
from kfp import dsl

def print_op(message: str):
    return dsl.ContainerOp(
        name='Print Message',
        image='alpine:3.6',
        command=['echo', message]
    )

def sum_op(a: int, b: int):
    return dsl.ContainerOp(
        name='Sum Numbers',
        image='python:3.7',
        command=['python', '-c'],
        arguments=['print({} + {})'.format(a, b)]
    )

@dsl.pipeline(
    name='Simple Pipeline',
    description='A simple pipeline that prints a message and sums two numbers.'
)
def simple_pipeline(message: str, a: int, b: int):
    task1 = print_op(message)
    task2 = sum_op(a, b)

# Compile pipeline
from kfp.compiler import Compiler
Compiler().compile(simple_pipeline, 'simple_pipeline.yaml')

# Submit pipeline
client = kfp.Client()
client.create_run_from_pipeline_func(
    simple_pipeline,
    arguments={'message': 'Hello, Kubeflow!', 'a': 5, 'b': 10}
)
```
